````python
from datetime import datetime, timedelta

# Crawl Data
all_keyword = [
    "board of piece indonesia", "bop indonesia", "prabowo board of piece", "board of piece 17 triliun"
]

# 6 month ago
# since = datetime(2025, 1, 1).strftime('%Y-%m-%d')
# until = datetime(2025, 12, 31).strftime('%Y-%m-%d')

since = datetime(2025, 1, 1)
until = datetime(2026, 5, 30)

print(f"Since: {since}")
print(f"Until: {until}")
````

````python
all_keyword = [keyword.lower() for keyword in all_keyword]

for i, keyword in enumerate(all_keyword):
  filename = f'{keyword}.csv'
  search_keyword = f'{keyword} since:{since} until:{until}'
  limit = 5000

  !npx --yes tweet-harvest -o "{filename}" -s "{search_keyword}" -l {limit} --token {twitter_auth_token}
  query_print = f"{search_keyword} since:{since} until:{until}"
  print(f"{i+1}. {query_print}")
````

Kode yang diberikan berfungsi untuk melakukan pengumpulan data tweet terkait beberapa kata kunci tentang ancaman keamanan siber dan kebocoran data di Indonesia dengan menggunakan **tweet-harvest**, yang merupakan sebuah alat untuk mengumpulkan tweet dari Twitter.

Berikut penjelasan rinci dari kode tersebut:

1. **List of Keywords** (`all_keyword`):
   - `all_keyword` adalah sebuah list yang berisi beberapa kata kunci yang berkaitan dengan ancaman keamanan siber dan kebocoran data di Indonesia.
   - Semua kata kunci tersebut kemudian diubah menjadi huruf kecil dengan list comprehension:
     ```python
     all_keyword = [keyword.lower() for keyword in all_keyword]
     ```

2. **Looping untuk Mengumpulkan Data**:
   - `for i, keyword in enumerate(all_keyword)` adalah loop yang iterasi setiap elemen dari `all_keyword` dengan menggunakan `enumerate` untuk mendapatkan indeks dan nilai dari setiap kata kunci.
   - `i` adalah indeks (nomor urut) dari kata kunci, sementara `keyword` adalah kata kunci itu sendiri dalam format huruf kecil.

3. **Penentuan Nama File**:
   - Untuk setiap kata kunci, kode membuat nama file CSV berdasarkan kata kunci tersebut. Misalnya, jika kata kunci adalah "ancaman keamanan siber", nama file yang dibuat adalah `ancaman keamanan siber.csv`.
     ```python
     filename = f'{keyword}.csv'
     ```

4. **Perintah untuk Mencari Tweet**:
   - `search_keyword` diatur sama dengan nilai `keyword`, yaitu kata kunci yang digunakan untuk pencarian tweet di Twitter.
   - `limit = 5000` adalah batasan jumlah tweet yang ingin diambil, yaitu sebanyak 5000 tweet.

5. **Pengumpulan Data Menggunakan `tweet-harvest`**:
   - Perintah `!npx --yes tweet-harvest -o "{filename}" -s "{search_keyword}" -l {limit} --token {twitter_auth_token}` digunakan untuk menjalankan `tweet-harvest` (alat pengumpul data tweet).
     - `npx`: Menjalankan perintah tanpa perlu menginstal alat terlebih dahulu.
     - `--yes`: Melewati konfirmasi yang biasanya diperlukan.
     - `tweet-harvest`: Nama alat yang digunakan untuk mengumpulkan tweet.
     - `-o "{filename}"`: Menyimpan hasil pencarian tweet ke dalam file dengan nama yang sesuai dengan kata kunci.
     - `-s "{search_keyword}"`: Kata kunci yang digunakan untuk pencarian tweet.
     - `-l {limit}`: Batasan jumlah tweet yang ingin dikumpulkan (5000 tweet).
     - `--token {twitter_auth_token}`: Token akses Twitter yang diperlukan untuk melakukan pencarian tweet.

Dengan kode ini, untuk setiap kata kunci yang ada dalam `all_keyword`, program akan mencari tweet terkait dan menyimpan hasilnya dalam file CSV dengan nama sesuai kata kunci.

````python
import pandas as pd
import os

csv_dir = "tweets-data"
all_csv_paths = [os.path.join(csv_dir, file_) for file_ in os.listdir(csv_dir) if file_.endswith('.csv')]

# Filter out empty files before attempting to read them
valid_csv_files = [file_path for file_path in all_csv_paths if os.path.getsize(file_path) > 0]

# excel file
excel_file = "BoP Indonesia_X_Scrapping.xlsx"

# Read only valid (non-empty) CSV files into a list of DataFrames
all_df = []
for file_path in valid_csv_files:
    try:
        df_temp = pd.read_csv(file_path)
        if not df_temp.empty: # Ensure the DataFrame is not empty after reading (e.g., only header)
            all_df.append(df_temp)
    except pd.errors.EmptyDataError: # Catch if file was empty or only headers
        print(f"Skipping empty or malformed file: {file_path}")

# Check if there are any dataframes to concatenate
if all_df:
    df = pd.concat(all_df)
    df = df.reset_index(drop=True)
    df = df.drop_duplicates()
    # Ensure 'full_text' column exists before selecting it
    # if 'full_text' in df.columns:
    #     df = df[['full_text']]
    # else:
    #     print("Warning: 'full_text' column not found in concatenated data. Exporting all available columns.")

    df.to_excel(excel_file, index=False)
    # The 'filename' variable might not be defined if the loop in zhfAzr2oQdi0 didn't complete or if this cell is run independently.
    # It's better to use a fixed name or ensure 'filename' is correctly set if intended for specific use.
    # For now, commenting out the second excel export to avoid potential NameError if 'filename' is not guaranteed to be the last one.
    # df.to_excel(filename.replace('.csv', '.xlsx'), index=False)
else:
    print("No valid data to process from CSV files after filtering. An empty Excel file will be created.")
    df = pd.DataFrame(columns=['full_text']) # Create an empty DataFrame to avoid errors in subsequent steps if any
    df.to_excel("kebijakan menteri esdm dalam mempertahankan harga bbm bersubsidi_X_Scrapping.xlsx", index=False)
````

Kode yang diatas bertujuan untuk menggabungkan beberapa file CSV yang berisi data tweet, menghapus duplikat, dan kemudian menyimpan hasilnya dalam format Excel. Berikut adalah penjelasan rinci untuk setiap bagian kode:

1. **Impor Library**:
   ```python
   import pandas as pd
   import os
   ```
   - `pandas` diimpor untuk memproses data dalam bentuk DataFrame, yang merupakan struktur data utama untuk mengolah data tabular.
   - `os` diimpor untuk bekerja dengan sistem berkas, terutama untuk membaca daftar file di dalam direktori.

2. **Menentukan Direktori CSV**:
   ```python
   csv_dir = "tweets-data"
   ```
   - Variabel `csv_dir` diatur ke nama direktori tempat file-file CSV yang berisi data tweet disimpan, yaitu dalam folder "tweets-data".

3. **Mengambil Daftar File CSV**:
   ```python
   all_csv = [csv_dir+'/'+file_ for file_ in os.listdir(csv_dir)]
   ```
   - `os.listdir(csv_dir)` menghasilkan daftar nama file yang ada dalam folder `tweets-data`.
   - List comprehension digunakan untuk membentuk daftar `all_csv`, yang berisi path lengkap ke setiap file CSV di dalam direktori tersebut.

4. **Membaca Setiap File CSV ke dalam DataFrame**:
   ```python
   all_df = [pd.read_csv(file_) for file_ in all_csv]
   ```
   - Untuk setiap file CSV yang ada dalam `all_csv`, file tersebut dibaca menggunakan `pd.read_csv(file_)` yang akan mengembalikan DataFrame.
   - List comprehension digunakan untuk membuat daftar `all_df`, yang berisi DataFrame untuk setiap file CSV.

5. **Menggabungkan Semua DataFrame**:
   ```python
   df = pd.concat(all_df)
   ```
   - `pd.concat(all_df)` digunakan untuk menggabungkan semua DataFrame yang ada dalam `all_df` menjadi satu DataFrame tunggal, `df`.

6. **Mereset Indeks**:
   ```python
   df = df.reset_index(drop=True)
   ```
   - `reset_index(drop=True)` digunakan untuk mereset indeks dari DataFrame yang telah digabungkan. Dengan `drop=True`, indeks lama akan dibuang dan tidak ditambahkan sebagai kolom baru.

7. **Menghapus Duplikat**:
   ```python
   df = df.drop_duplicates()
   ```
   - `df.drop_duplicates()` digunakan untuk menghapus baris yang duplikat dalam DataFrame, sehingga hanya data yang unik yang tersisa.

8. **Menyimpan ke dalam File Excel**:
   ```python
   df.to_excel("Ancaman Keamanan Siber dan Kebocoran Data dalam Pilkada Sentimen.xlsx", index=False)
   ```
   - `df.to_excel()` digunakan untuk menyimpan DataFrame `df` ke dalam file Excel dengan nama `"Ancaman Keamanan Siber dan Kebocoran Data dalam Pilkada Sentimen.xlsx"`.
   - `index=False` memastikan indeks tidak disertakan dalam file Excel.

9. **Menyimpan ke File Excel Berdasarkan Nama File CSV**:
   ```python
   df.to_excel(filename.replace('.csv', '.xlsx'), index=False)
   ```
   - Baris ini menyimpan DataFrame `df` ke dalam file Excel dengan nama yang sama dengan file CSV asli, namun dengan ekstensi `.xlsx`.
   - `filename.replace('.csv', '.xlsx')` mengubah nama file CSV yang bersangkutan menjadi nama file Excel.

### Kesimpulan:
Kode ini pertama-tama menggabungkan beberapa file CSV yang berisi data tweet, kemudian membersihkan data dengan menghapus duplikat, mereset indeks, dan akhirnya menyimpan hasilnya dalam format Excel dengan dua cara: satu dengan nama file tetap ("Ancaman Keamanan Siber dan Kebocoran Data dalam Pilkada Sentimen.xlsx") dan satu lagi dengan nama yang diambil dari file CSV asli (mengganti ekstensi `.csv` menjadi `.xlsx`).
